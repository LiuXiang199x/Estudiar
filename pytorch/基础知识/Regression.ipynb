```python
import torch
from torch.autograd import Variable
import torch.nn.functional as F
import matplotlib.pyplot as plt
```

```python
x = torch.unsqueeze(torch.linspace(-1, 1, 100), dim=1)  # x data (tensor), shape=(100, 1)
y = x.pow(2) + 0.2*torch.rand(x.size())                 # noisy y data (tensor), shape=(100, 1)

x, y = Variable(x), Variable(y)

plt.scatter(x.data.numpy(), y.data.numpy())
plt.show()
```

```python
# 一个网络中必须的是init和forward两个部分。
# init：搭建所有层的信息。forward：定义了前向传递的过程。

class Net(torch.nn.Module):
    def __init__(self, n_feature, n_hidden, n_output):
        super(Net, self).__init__()
        self.hidden = torch.nn.Linear(n_feature, n_hidden)   # hidden layer
        self.predict = torch.nn.Linear(n_hidden, n_output)   # output layer

    def forward(self, x):
        # print(x.size())   # [100,1]  
        # 一个神经元输入，[100,1]，隐藏层十个神经元，十个参数w[10,1]和十个bias[10]
        
        x = F.relu(self.hidden(x))      # activation function for hidden layer
        # 十个神经元的输出：
        # print(x.size())  # [100, 10]
        
        x = self.predict(x)             # linear output
        # print(x.size())  # [100, 1]
        return x

net = Net(n_feature=1, n_hidden=10, n_output=1)     # define the network
print(net)  # net architecture
print(len(list(net.parameters()))) # 4
for param in list(net.parameters()):
    print(param.size())

"""
一个输入到十个神经元，十个w，然后还有十个bias。 接着从十个神经元到一个输出。一个bias
torch.Size([10, 1])
torch.Size([10])
torch.Size([1, 10])
torch.Size([1])
"""
    
optimizer = torch.optim.SGD(net.parameters(), lr=0.2)
loss_func = torch.nn.MSELoss()  # this is for regression mean squared loss

plt.ion()   # something about plotting

for t in range(200):
    prediction = net(x)     # input x and predict based on x

    loss = loss_func(prediction, y)  # pred在前，y真实值在后
    
    # 三个优化步骤,因为optmizer是要优化网络中所有的参数
    # 所以先将参数的所有梯度降为零，因为每次算完loss更新后它的梯度都会保留在参数中
    optimizer.zero_grad()
    loss.backward()   # 计算梯度
    optimizer.step()   # 以0.5步长来优化梯度

    if t % 5 == 0:
        # plot and show learning process
        plt.cla()
        plt.scatter(x.data.numpy(), y.data.numpy())
        plt.plot(x.data.numpy(), prediction.data.numpy(), 'r-', lw=5)
        plt.text(0.5, 0, 'Loss=%.4f' % loss.data.numpy(), fontdict={'size': 20, 'color':  'red'})
        plt.pause(0.1)

plt.ioff()
plt.show()
```
